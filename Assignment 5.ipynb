요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> print c > 10 
True 

> print d > 10 
False 

> print c < 10 
False 

> print d < 10 
True 

> print c == 17 
True 

> print d != 9 
False 

```python
class Counter:
    def __init__(self, count ,step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count (step : %d)] %d" % (self.step, self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        return self.count + other
    def __sub__(self, other):
        return self.count - other
    def __it__(self, other):
        return self.count < other
    def __le__(self, other):
        return self.count <= other
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other
    def __gt__(self, other):
        return self.count > other
    def __ge__(self, other):
        return self.count >= other

c = Counter(10)
d = Counter(10, 2)

print c
print d
print
c.incr()
d.incr()

print c
print d
print

c()
d()

print c
print d
print

c = c + 5
d = d - 5

print c
print d
print

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

init() 메소드는 클래스 객체가 생성될때 자동으로 실행되는 메소드로서 클래스 내에 init() 메소드가 없는 경우는 기본설정된 init()메소드가 자동으로 호출되고 따로 정의를 해둔 경우는 정의된 init() 메소드가 실행된다. 보통 __  __로 구성된 메소드들은 기본적으로 구현이 되어있고 특정 상황에서 자동으로 실행되는 함수들이다.일반적으로는 기본 함수를 쓰되 필요에 따라서는 재정의를 해서 쓰기도 한다.
위의 코드에서는 init() 메소드의 인자로 자기 자신을 참조할 수 있는 self, 그리고 인자 l을 가지고 있다. MySet이 현재 list 클래스를 상속 받고 있으므로 l에는 리스트가 들어가는 것으로 생각하면 될 것 같다. init의 코드를 보면 처음에 반복문을 통해 l에서 원소들을 가져와 append를 통해 모두 저장한 뒤 정적 메소드인 eliminate_duplicate를 수행하고 끝나게 된다. 
eliminate_duplicate 메소드는 정적 메소드로서 클래스의 내부에 있으나 일반 메소드와 달리 인스턴스 객체와는 상관없는 메소드로서 함수 인자에 self를 받지 않는것이 특징이다. 또한 인스턴스 이름이 아닌 클래스 이름으로 부르는 것이 특징이다. 인스턴스 이름으로도 부르는 것이 가능하지만 init() 메소드에선 인스턴스 객체가 없으므로 클래스 이름으로 부르고 인자로는 self를 받는다. 
eliminate_duplicate 메소드는 l을 받아 l에서 중복된 값이 제거된 결과를 저장할 s를 리스트 자료형으로 선언하고 반복문을 통해 l의 인자들 중 s에 없을 경우에만 s에 추가하도록 한다. 그 뒤 지역변수 l[:]을 선언해서 이곳에 s의 인자들을 다시 저장한다.
str() 메소드는 print 함수가 인스턴스를 출력할때 자동으로 실행되는 메소드이다. result에 특정 문자열을 넣어두고 반복문을 통해 클래스에 있는 저장값들을 다시 문자로 변환하여 result에 붙여준다. 그 뒤 마지막 숫자 뒤의 "," 기호를 빼고 "}"를 넣은 뒤 저장된 result를 반환하여 print 함수가반환된 result를 출력하도록 한다.

[5번 문제]
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> u = s | t 

> print u 

MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> u = s & t 

> print u 

MySet: {2, 3} 

요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> s = MySet([1, 2, 3]) 

> t = MySet([3, 4, 5]) 

> u = s - t 

> print u 

> MySet: {1,2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    def __or__(self, other):
        #합집합
        for e in other:
            if e not in self:
                self.append(e)
        return self
    def __and__(self, other):
        #교집합
        for e in self:
            if e not in other:
                self.remove(e)
        return self
                
    def __sub__(self, other):
        #차집합
        for e in other:
            if e in self:
                self.remove(e)
        return self

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s | t
    print u
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s & t
    print u
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print u
```

```python
sampleFile=open("18.txt", "r")

triangle=[]

for l in sampleFile.readlines():
    l=l.rstrip('\n')
    raw_list=l.split(" ")
    
    temp_l=list()
    for y in raw_list:
        temp_l.append(int(y))
    
    triangle.append(temp_l)
    
# reverse.
triangle.reverse();

# add each row to row+1 and replace row
for row in range(0,14):
    for col in range(0, 14-row):
        first_sum = triangle[row][col] + triangle[row+1][col]
        second_sum = triangle[row][col+1] + triangle[row+1][col]
        max_of_two = max(first_sum, second_sum)
        
        triangle[row+1][col] = max_of_two
        

print triangle[14]
```

[12번 문제]

프로젝트 오일러 문제 21

```python
import math

def divi(n):
    cnt = [] #약수를 저장하기 위한 리스트 
    setting = int(math.sqrt(n)) # n의 약수를 더 빠르게 구하기 위한 기준값
    for x in range(1, setting+1):
        if(n % x == 0): #나누어 질 경우
            cnt.append(x)
    for i in range(0, len(cnt)):
        if(n / cnt[i] == cnt[i]): #제곱근이 있는 경우 이미 리스트에 있으므로 반복문의 상단으로 올라감
            continue
        cnt.append(n/cnt[i])
        
    cnt.remove(n) #자기 자신은 지움
    return sum(cnt)

n = 2
result = 0
while n < 10001: 
    if(n == divi(divi(n)) and n != divi(n)): #친화수의 조건, 쌍이어야 하므로 n 과 약수의 합이 같으면 쌍이 될 수 없다
        result += n
    n += 1

print result
```

```python
# -*- coding: utf-8 -*-
import urllib2
import string

class WebWordsFrequency:
    def __init__(self, *url):
        self.l = list()
        url = list(url)
        for u in range(0, len(url)):
            self.l.append(url[u])
    def addUrl(self, url):
        if(self.l.count(url) == 0):
            self.l.append(url)
        elif(self.l.count(url) > 0):
            print "Already exist."
    def removeUrl(self, url):
        if(self.l.count(url) != 0):
            self.l.remove(url)
        else:
            print "Url don't exist.(removeUrl's result)"
    def listUrl(self):
        if (len(self.l) == 0):
            print "Url don't exist.(listUrl's result)"
        else:
            for u in self.l:
                print u
    def getWordsFrequency(self):
        #메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환하시오.
        #만약 등록된 URL이 없다면 공백 사전을 반환
        words_dic = dict() # 사전으로 담기 위한 변수
        eng_list = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into',
            'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the',
            'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
        if (len(self.l) == 0):
            return words_dic
        else:
            for url in self.l:
                source = urllib2.urlopen(url).read()
        
                for i in range(0, source.count('<!--')):
                    if source.find("<!--") != -1:
                        source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")
    
                for i in range(0,source.count("<")):
                    if source.find("<") != -1:
                        source = source.replace(source[source.find("<"):source.find(">")+1:],"")
    
                for punc in string.punctuation: # 구두문자를 인자로 반복문 실행
                    source = source.replace(punc, "") #source에서 구두문자를 공백으로 바꾼 것을 다시 저장
        
                words = source.split() # 공백으로 단어를 나누어 리스트 저장
    
                #영어 불용자 제거, 불용자는 단어이므로 위의 구두문자와는 다르게 단어를 나눈뒤 삭제
                p = 0
                while p == 0:
                    for e in eng_list:
                        if(e in words): #영어 불용자가 리스트에 있을경우 탐색을 멈춤
                            break
                        else: # 영어 불용자가 리스트에 없는 경우 무한 루프 종료
                            p = 1
                            continue #리스트에 영어 불용자가 없으므로 아래의 반복문 무시
         
                    #영어 불용자 삭제
                    for e in eng_list:
                        if(e in words):
                            words.remove(e)

                inner_words_dic = dict()
                for word in words: #단어들을 인자로 반복문 실행
                    inner_words_dic[word] = words.count(word) # 단어를 키로 단어의 갯수를 값으로 사전에 저장
                
                #중복단어들을 찾아 빈도수를 더해주고 반환할 사전에 업데이트
                for i in inner_words_dic:
                    for w in words_dic:
                        if(i == w):
                            inner_words_dic[i] += words_dic[w]
                words_dic.update(inner_words_dic)
            
            return words_dic
        
    def getMaxFreqencyWords(self):
        
        def cmp_val(a1, a2): #빈도수를 정렬하기 위한 함수
            return cmp(a1[1], a2[1])

        words_dic = self.getWordsFrequency()
        if(len(words_dic) == 0): #getWordsFrequency() 에서 받아온 사전이 아무것도 가지지 않을 경우
            return None
        else:
            max_list = list() #빈도수가 가장 높은 것들을 가지고 반환할 리스트
            #사전에서 리스트로 만들고 내림차순으로 정렬 후 비교
            words_dic_list = words_dic.items()
            words_dic_list.sort(cmp_val, reverse = True)
            k_max = words_dic_list[0][1] #내림차순 비교이므로 가장 앞에 있는 튜플의 value값이 최대빈도수
            max_list.append(words_dic_list[0][0]) #내림차순 비교이므로 가장 앞에 있는 단어를 넣는다
            #같은 빈도수의 단어를 확인
            for w in range(1, len(words_dic_list)):
                if(words_dic_list[w][1] ==  k_max): #같은 빈도수가 있을경우
                    max_list.append(words_dic_list[w][0]) #단어 추가
                elif(words_dic_list[w][1] <  k_max): #최대빈도수 보다 작은 경우 
                    break #반복문을 바로 빠져나옴
            return max_list #최대 빈도수 단어가 들어있는 리스트 반환

class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self, *url):
        WebWordsFrequency.__init__(self, *url)
        self.index = 0 #next 메소드에서 배열에 쓸 인덱스 변수
        self.global_words_dic_list = self.getWordsFrequency() #url을 분석해 나온 단어와 빈도가 들어있는 리스트
    def addUrl(self, url):
        WebWordsFrequency.addUrl(self, url)
        self.global_words_dic_list = self.getWordsFrequency() #url이 추가되는 경우 다시 리스트를 가져온다
    def removeUrl(self, url):
        WebWordsFrequency.removeUrl(self, url)
        self.global_words_dic_list = self.getWordsFrequency() #url이 삭제되는 경우 다시 리스트를 가져온다
    def __iter__(self):
        return self
    def next(self):
        if(len(self.global_words_dic_list) == self.index): #인덱스의 숫자와 리스트의 크기가 같은경우
            raise StopIteration 
            self.index = 0 #인덱스를 0으로 초기화
        else:
            words_next = self.global_words_dic_list[self.index] #반환할 리스트의 내용
            self.index += 1 #인덱스 증가
        return words_next #단어와 빈도의 튜플을 반환
        
    def getWordsFrequency(self, reverse = False):
        def cmp_val(a1, a2): #빈도수를 정렬하기 위한 함수
            return cmp(a1[1], a2[1])
        words_dic = WebWordsFrequency.getWordsFrequency(self) #빈도수가 담긴 사전을 가지고 옮
        words_dic_list = words_dic.items() #사전을 리스트로 바꿈
        #위의 reverse 인자에 따라 오름차순 내림차순으로 정렬
        #위의 reverse 인자에 따라 아래의 식에서 반대로 넣어주어야 원하는 차순 정렬이 된다
        if (reverse == False):
            words_dic_list.sort(cmp_val, reverse = True)
        else:
            words_dic_list.sort(cmp_val, reverse = False)
        return words_dic_list
    
    def getMaxFreqencyWords(self):
        
        def cmp_val(a1, a2): #빈도수를 정렬하기 위한 함수
            return cmp(a1[1], a2[1])

        words_dic_list = self.getWordsFrequency()
        if(len(words_dic_list) == 0): #getWordsFrequency() 에서 받아온 사전이 아무것도 가지지 않을 경우
            return None
        else:
            max_list = list() #빈도수가 가장 높은 것들을 가지고 반환할 리스트
            #사전에서 리스트로 만들고 내림차순으로 정렬 후 비교
            words_dic_list.sort(cmp_val, reverse = True)
            k_max = words_dic_list[0][1] #내림차순 비교이므로 가장 앞에 있는 튜플의 value값이 최대빈도수
            max_list.append(words_dic_list[0][0]) #내림차순 비교이므로 가장 앞에 있는 단어를 넣는다
            #같은 빈도수의 단어를 확인
            for w in range(1, len(words_dic_list)):
                if(words_dic_list[w][1] ==  k_max): #같은 빈도수가 있을경우
                    max_list.append(words_dic_list[w][0]) #단어 추가
                elif(words_dic_list[w][1] <  k_max): #최대빈도수 보다 작은 경우 
                    break #반복문을 바로 빠져나옴
            return max_list #최대 빈도수 단어가 들어있는 리스트 반환

            

w1 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency()
 
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')

w1.listUrl()
print "=" * 10
w2.listUrl()
print "=" * 10
w3.listUrl()
print "=" * 10

print w1.getWordsFrequency() 
```
